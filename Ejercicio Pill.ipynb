```python
from PIL import Image
```

```python
fName="fletcher.png"
```

```python
img=Image.open(fName)
```

```python
print(img.format)
print(img.mode)
print(img.size)
```

```python
#Depicting image

img.show()
```

```python
#Show data values distribution

#Como se ve en el script la fotografia que elegi tenia el mode RGBA entonces para representar la escala quito 
#la variable alpha y asi se representa como en el ejemplo proporcionado

import matplotlib.pyplot as plt 

img_g= img.convert('RGB')

h= img_g.histogram()

plt.plot(h)
```

```python
#For color images, the histograms for all bands are concatenated
```

```python
#Changing image size
```

```python
#Preserving aspect ratio

img.thumbnail((1200,800))
plt.imshow(img)

print(img.size)
```

```python
#Without reserving aspect radio

img_1=img.resize((600,400))
print(img_1.size)

plt.imshow(img_1)
```

```python
# Simple image manipulations functions

img_t=img.transpose(Image.FLIP_TOP_BOTTOM)
print(img_t.size)

plt.imshow(img_t)
```

```python
# Rotating

img_r=img.rotate(45)
print(img_r.size)

plt.imshow(img_r)
```

```python
# Cropping 

img_c=img.crop((400,400,600,600))
print(img_c.size)

plt.imshow(img_c)
```

```python
# Saving image

img_c.save("lollipop.png")
```
